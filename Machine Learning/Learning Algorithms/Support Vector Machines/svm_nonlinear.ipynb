```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
'''Loading and Storing dataset values in X and y'''
data = pd.read_csv('nonlinear_svm_data.csv')
X = np.array(data[['x1', 'x2']].values)
y = np.array(data['label'].values)
```

```python
'''
Visualizing the dataset
'''
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Non Linear SVM Data Visualization')
plt.show()
```
