```python
import numpy as np
import pandas as pd

np.random.seed(42)

N = 100_000   # size of dataset

# Class 0
X0 = np.random.randn(N//2, 2) + np.array([-3, -3])
y0 = np.zeros(N//2)

# Class 1
X1 = np.random.randn(N//2, 2) + np.array([3, 3])
y1 = np.ones(N//2)

X = np.vstack((X0, X1))
y = np.hstack((y0, y1))

df = pd.DataFrame(X, columns=['x1', 'x2'])
df['label'] = y.astype(int)

df.to_csv("linear_svm_data.csv", index=False)

print(df.head())
print("Saved linear_svm_data.csv")
```
