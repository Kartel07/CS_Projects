```python
from sklearn.datasets import make_moons
import pandas as pd

X, y = make_moons(
    n_samples=100_000,
    noise=0.2,
    random_state=42
)

df = pd.DataFrame(X, columns=['x1', 'x2'])
df['label'] = y

df.to_csv("nonlinear_svm_data.csv", index=False)

print(df.head())
print("Saved nonlinear_svm_data.csv")
```
