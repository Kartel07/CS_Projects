```python
import numpy as np
import pandas as pd
np.random.seed(42)

N = 50_000
D = 50   # high dimensional

X = np.random.randn(N, D)

# True linear separator
true_w = np.random.randn(D)
y = (X @ true_w > 0).astype(int)

df = pd.DataFrame(X, columns=[f'f{i}' for i in range(D)])
df['label'] = y

df.to_csv("high_dim_svm_data.csv", index=False)

print("Saved high_dim_svm_data.csv")
```
