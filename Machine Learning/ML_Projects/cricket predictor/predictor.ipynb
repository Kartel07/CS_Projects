```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('IPL_Prematch.csv', low_memory=False)
```

Data Cleaning and Preparation

- It was completed using Excel and SQL, original unclean and clean datasets are included in the folder


Data visualisation


```python
df = pd.DataFrame(data)

#deleting null values
df.dropna(inplace=True)
#deleting duplicates
df = df.drop_duplicates(keep = 'first')

df.head()
```

```python
sns.pairplot(df)
plt.show()
```

```python
co_mtx = data.corr(numeric_only=True)

# Print correlation matrix
print(co_mtx)

# Plot correlation heatmap
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)

# Display heatmap
plt.show()
```
